### Train Encoder-Decoder

Load sentences

```python
import json

def load_sentences(data_path):
    with open(data_path, "r") as f:
        data = json.load(f)
    sentences = []
    for step, s_data in data.items():
        if not step.startswith("Step"):
            continue
        sentences.append(s_data["Agent_0"]["Sentence"][1:-1])
        sentences.append(s_data["Agent_1"]["Sentence"][1:-1])
    return sentences
sentences = load_sentences("test_data/Sentences_Generated_P1.json")
```

Initialise Encoder and Decoder

```python
import torch
from model.modules.lm import GRUEncoder, GRUDecoder, OneHotEncoder

word_encoder = OneHotEncoder(
    ['South','Not','Located','West','Object','Landmark','North','Center','East'])

enc = GRUEncoder(32, word_encoder)
dec = GRUDecoder(32, word_encoder)
```

Initialise Loss and Optimizer

```python
from torch import nn, optim

criterion = nn.NLLLoss()
optim = optim.SGD(list(enc.parameters()) + list(dec.parameters()), lr=0.001)
```

Train

```python
import time
from tqdm import tqdm 

def train(sentences, enc, dec, word_encoder, criterion, optim):
    start = time.time()
    
    plot_losses = []
    print_loss_total = 0  # Reset every print_every
    
    for s_i in tqdm(range(len(sentences))):
        optim.zero_grad()
        
        loss = 0
        
        # Encoder forward pass
        enc_hidden_state = enc([sentences[s_i]])
        
        # Decoder forward pass
        enc_hidden_state = enc_hidden_state.squeeze(0)
        print(enc_hidden_state.shape)
        encoded_targets = word_encoder.encode_batch([sentences[s_i]])
        decoder_outputs, _ = dec(enc_hidden_state, encoded_targets)
        
        # Compute loss
        for d_o, e_t in zip(decoder_outputs, encoded_targets):
            e_t = e_t.long()
            for w_i in range(d_o.size(0)):
                # print("LOSS", d_o[w_i], e_t[w_i], criterion(d_o[w_i], e_t[w_i]))
                loss += criterion(d_o[w_i], e_t[w_i])
        
        # Backprop
        # print(loss)
        loss.backward()
        optim.step()
        
        plot_losses.append(loss.item() / len(sentences[s_i]))
    
    return plot_losses
            
plot_loss = train(sentences, enc, dec, word_encoder, criterion, optim)
```

```python
import matplotlib.pyplot as plt

plt.plot(plot_loss)
```

```python
import time
from tqdm import tqdm 

def train(sentences, enc, dec, word_encoder, criterion, optim, batch_size=1):
    start = time.time()
    
    plot_losses = []
    print_loss_total = 0  # Reset every print_every
    
    for s_i in tqdm(range(0, len(sentences), batch_size)):
        optim.zero_grad()
        
        loss = 0
        
        # Encoder forward pass
        enc_hidden_state = enc(sentences[s_i:s_i + batch_size])
        
        # Decoder forward pass
        enc_hidden_state.squeeze(0)
        encoded_targets = word_encoder.encode_batch(sentences[s_i:s_i + batch_size])
        decoder_outputs, _ = dec(enc_hidden_state.squeeze(0), encoded_targets)
        
        # Compute loss
        tot_len = 0
        for d_o, e_t in zip(decoder_outputs, encoded_targets):
            e_t = e_t.long()
            for w_i in range(d_o.size(0)):
                tot_len
                loss += criterion(d_o[w_i], e_t[w_i])
        
        # Backprop
        loss.backward()
        optim.step()
        
        tot_len = sum([s.size(0) for s in encoded_targets])
        plot_losses.append(loss.item() / tot_len)
    
    return plot_losses
            
plot_loss = train(sentences, enc, dec, word_encoder, criterion, optim)
```
