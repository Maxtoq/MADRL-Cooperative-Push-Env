```python
from src.utils.decay import ParameterDecay
import matplotlib.pyplot as plt

comm_eps = ParameterDecay(1.0, 0.01, 10000, "sigmoid", )

e = [
    comm_eps.get_explo_rate(i)
    for i in range(10000)]

plt.plot(e)
```

```python
for s in [0.5, 0.8, 1, 2, 3, 5, 8]:
    comm_eps = ParameterDecay(1.0, 0.01, 10000, "sigmoid", s)

    e = [
        comm_eps.get_explo_rate(i)
        for i in range(10000)]

    plt.plot(e, label=str(s))
plt.legend()
```
