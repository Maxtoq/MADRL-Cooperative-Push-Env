# Analyse a single file

For each batch of steps, compute these metrics:
* `n_perf_sent`: Number of different perfect sentence
* `n_gen_sent`: Number of different generated sentence
* `ratio_perf_gen`: Ratio of perfect sentences that have been generated (are also in generated sentences)
* `mean_len_perf`: Average length of perfect sentences
* `mean_len_gen`: Average length of generated sentences
* `ratio_gen_is_perf`: Ratio of generated sentences that are identical to the corresponding perfect sentences
* `mean_n_same_token`: Average number of similar tokens between generated and perfect sentences
* `mean_ratio_same_token`: Average ratio of similar tokens between generated and perfect sentences

```python
def n_same_token(s1, s2):
    return len(set(s1) & set(s2))

def get_file_info(file_path):
    df = load_file(file_path)
    
    info = {
        "n_perf_sent": df["Perfect Message"].nunique(),
        "n_gen_sent": df["Generated Message"].nunique()
    }
    
    info["ratio_perf_gen"] = (
        np.array([len(df.loc[df["Generated Message"] == x]) 
                  for x in df["Perfect Message"].unique()]) 
        > 0).mean()
    
    def remove_empty(s):
        if '' in s:
            s.remove('')
        return s
    lang_columns = ["Generated Message", "Perfect Message", "Broadcasted Message"]
    for c in lang_columns:
        df[c] = df[c].apply(lambda x: remove_empty(x.split(" ")))
        
    df["len_perfect"] = df["Perfect Message"].apply(len)
    df["len_generated"] = df["Generated Message"].apply(len)
    
    info["mean_len_perf"] = df["len_perfect"].mean()
    info["mean_len_gen"] = df["len_generated"].mean()
    
    df["gen_is_perf"] = df["Generated Message"] == df["Perfect Message"]
    info["ratio_gen_is_perf"] = df["gen_is_perf"].sum() / len(df)
    
    df["n_same_token"] = df.apply(lambda x: n_same_token(x["Generated Message"], x["Perfect Message"]), axis=1)
    info["mean_n_same_token"] = df["n_same_token"].mean()
    
    df["ratio_same_token"] = df["n_same_token"] / df["len_generated"]
    info["mean_ratio_same_token"] = df["ratio_same_token"].dropna().mean()
    
    return info
```

# Print all in one

```python
import re
import os
import numpy as np
from tqdm import tqdm
import pandas as pd
import matplotlib.pyplot as plt


def hist_sent_occ(file_path):
    df = pd.read_csv(file_path)
    
    df = df.fillna("")
    df.head()
    
    sent_occ = pd.DataFrame(columns=["Sentences", "Perfect N#", "Generated N#"])

    sent_occ["Sentences"] = pd.unique(df[["Perfect Message", "Generated Message"]].values.ravel('K'))

    for s in pd.unique(df[["Perfect Message", "Generated Message"]].values.ravel('K')):
        sent_occ.loc[sent_occ["Sentences"] == s, "Perfect N#"] = len(df[df["Perfect Message"] == s])
        sent_occ.loc[sent_occ["Sentences"] == s, "Generated N#"] = len(df[df["Generated Message"] == s])

    sent_occ = sent_occ.sort_values(["Perfect N#", "Generated N#"])#, ascending=[False, False])
    
    fig, ax = plt.subplots(figsize=(10, 40))

    sent_occ.plot(x="Sentences", y=["Perfect N#", "Generated N#"], kind="barh", ax=ax, width=1)

    for container in ax.containers:
        ax.bar_label(container, fontsize=7)

    plt.yticks(fontsize=7)
    plt.ylim(-0.5,len(sent_occ)-.5)

    ax.set_xscale('log')

def load_file(file_path):
    data = pd.read_csv(file_path)
    data = data.fillna("")
    return data

def get_step(file_name):
    r = re.search(r"cl\d*-(\d*).csv", file_name)
    return int(r.group(1))

def n_same_token(s1, s2):
    return len(set(s1) & set(s2))

def get_file_info(file_path):
    df = load_file(file_path)
    
    info = {
        "step": df["Step"]
        "n_perf_sent": df["Perfect Message"].nunique(),
        "n_gen_sent": df["Generated Message"].nunique()
    }
    
    info["ratio_perf_gen"] = (
        np.array([len(df.loc[df["Generated Message"] == x]) 
                  for x in df["Perfect Message"].unique()]) 
        > 0).mean()
    
    def remove_empty(s):
        if '' in s:
            s.remove('')
        return s
    lang_columns = ["Generated Message", "Perfect Message", "Broadcasted Message"]
    for c in lang_columns:
        df[c] = df[c].apply(lambda x: remove_empty(x.split(" ")))
        
    df["len_perfect"] = df["Perfect Message"].apply(len)
    df["len_generated"] = df["Generated Message"].apply(len)
    
    info["mean_len_perf"] = df["len_perfect"].mean()
    info["mean_len_gen"] = df["len_generated"].mean()
    
    df["gen_is_perf"] = df["Generated Message"] == df["Perfect Message"]
    info["ratio_gen_is_perf"] = df["gen_is_perf"].sum() / len(df)
    
    df["n_same_token"] = df.apply(lambda x: n_same_token(x["Generated Message"], x["Perfect Message"]), axis=1)
    info["mean_n_same_token"] = df["n_same_token"].mean()
    
    df["ratio_same_token"] = df["n_same_token"] / df["len_generated"]
    info["mean_ratio_same_token"] = df["ratio_same_token"].dropna().mean()
    
    return info

def printall(dir_path):
    sorted_file_list = sorted(os.listdir(dir_path), key=get_step)
    
    info_list = []
    for f in tqdm(sorted_file_list):
        path = os.path.join(dir_path, f)
        if os.path.isfile(path):
            info_list.append(get_file_info(path), ignore_index=True)
    info_df = pd.DataFrame(info_list)
            
    fig, axs = plt.subplots(1, 3, figsize=(15, 4))        
    
    info_df[["n_perf_sent", "n_gen_sent"]].plot(
        title="Diversity of sentences", 
        xlabel="x25000 training steps", 
        ylabel="Number of different sentences",
        ax=axs[0])
    info_df[["ratio_gen_is_perf", "mean_n_same_token", "mean_ratio_same_token"]].plot(
        title="Similarity between perfect and generated sentences", 
        xlabel="x25000 training steps",
        ax=axs[1])
    info_df[["mean_len_perf", "mean_len_gen"]].plot(
        title="Mean Message Length", 
        xlabel="x25000 training steps", 
        ylabel="Number of tokens",
        ax=axs[2])
    plt.show()
    
    hist_sent_occ(os.path.join(dir_path, sorted_file_list[-1]))
```

```python
printall("../models/magym_PredPrey/FT_commencmappo_obsdist/run1/logs/comm_logs/")
```
